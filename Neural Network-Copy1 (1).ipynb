```python
import pandas as np
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sc 
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations=1000
xs=np.random.uniform(low=-10,high=10,size=(observations,1))
zs=np.random.uniform(low=-10,high=10,size=(observations,1))

inputs= np.column_stack((xs,zs))
```

```python
input_size=2
output_size=1
inputs
```

```python
print (inputs.shape)
```

```python
noise= np.random.uniform(-1,1,(observations,1))

target= 13*xs +7*zs - 12 + noise
print(target.shape)
```

```python
target = target.reshape(observations,)
fig= plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot (xs,zs,target)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('target')
ax.view_init (azim=100)
plt.show()
target= target.reshape(observations,1)
```

```python
init_range=0.1
```

```python
weight = np.random.uniform (-init_range, init_range, size=(2,1))

biases = np.random.uniform (-init_range, init_range, size=1)
```

```python
print(weight)
print(biases)
```

```python
learning_rate= 0.01
```

```python

for i in range(1,100):
    outputs= np.dot(inputs,weight) + biases
    delta = outputs - target
    
    loss = (np.sum(delta ** 2) /2 / observations)
    print(loss)

    
    delta_scaled= (delta/observations)
    
    weight= weight - learning_rate* np.dot(inputs.T, delta_scaled)
    biases= biases - learning_rate* np.sum(delta_scaled)
```

```python
print(weight,biases) 
```

```python
plt.plot(outputs,target)
plt.show()
```
